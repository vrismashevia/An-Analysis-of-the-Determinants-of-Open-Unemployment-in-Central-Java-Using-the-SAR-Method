#Pemodelan Regresi Linear Berganda untuk Menganalisis Hubungan Variabel

```python
import pandas as pd
import statsmodels.formula.api as sm
from sklearn.preprocessing import StandardScaler

# Baca data Excel
data = pd.read_excel('/content/data_kapsel.xlsx', nrows=36)

# Kolom numerik yang akan distandarisasi
numerical_columns = ['y', 'x1', 'x2', 'x3', 'x4', 'x5']
data_numeric = data[numerical_columns]

# Standarisasi data
scaler = StandardScaler()
data_standardized = pd.DataFrame(scaler.fit_transform(data_numeric), columns=numerical_columns)

# Tambahkan kembali kolom 'Wilayah' untuk referensi
data_standardized['Wilayah'] = data['Wilayah']

# Formula regresi linear berganda
formula = 'y ~ x1 + x2 + x3 + x4 + x5'

# Membangun model regresi linear berganda
model = sm.ols(formula, data=data_standardized).fit()

# Menampilkan hasil regresi
print(model.summary())

# Koefisien regresi
coefficients = model.params

# Nilai R-squared
r_squared = model.rsquared

# Menampilkan koefisien dan R-squared
print("\nKoefisien Regresi:")
print(coefficients)
print("R-squared:", r_squared)
```

#Uji Asumsi Klasik dalam Regresi Linear

```python
# Uji Asumsi Klasik
import statsmodels.stats.api as sms

# 2. Uji Heteroskedastisitas
# Uji Breusch-Pagan
from statsmodels.stats.diagnostic import het_breuschpagan
bp_test = het_breuschpagan(model.resid, model.model.exog)
labels = ['LM-Statistic', 'LM-Statistic p-value', 'F-Statistic', 'F-Test p-value']
print("\nUji Heteroskedastisitas (Breusch-Pagan):")
print(dict(zip(labels, bp_test)))


# 3. Uji Autokorelasi
# Uji Durbin-Watson
from statsmodels.stats.stattools import durbin_watson
dw_statistic = durbin_watson(model.resid)
print("Uji Autokorelasi (Durbin-Watson):")
print("DW-Statistic:", dw_statistic)


# 4. Uji Normalitas
# Uji Shapiro-Wilk
from scipy.stats import shapiro
shapiro_test = shapiro(model.resid)
print("Uji Normalitas (Shapiro-Wilk):")
print("Shapiro-Wilk Statistic:", shapiro_test.statistic)
print("P-value:", shapiro_test.pvalue)
```

1. Uji Heteroskedasitas :

Hipotesis:

H0 (Hipotesis Nol): Tidak ada heteroskedastisitas (varian residual tetap).

H1 (Hipotesis Alternatif): Ada heteroskedastisitas (varian residual tidak tetap).

Daerah Kritis  : H0 diterima jika p-value lebih besar dari 0.05

Keputusan :
Nilai p-value dari LM-Test (0.156) dan F-Test (0.162) lebih besar dari 0.05. Maka H0 dinyatakan tidak ditolak

Kesimpulan :
Berdasarkan hipotesis, ini menunjukkan bahwa tidak ada bukti signifikan adanya heteroskedastisitas dalam model, sehingga asumsi homoskedastisitas terpenuhi.

2. Uji Autokorelasi :

Hipotesis:

H0 (Hipotesis Nol): Tidak ada autokorelasi (residual tidak berkorelasi antar waktu).
H1 (Hipotesis Alternatif): Ada autokorelasi (residual berkorelasi antar waktu).

Daerah Kritis : H0 diterima jika statistik Durbin-Watson berada di sekitar 2

Keputusan :
Nilai DW-Statistic (2.006) mendekati 2, sehingga H0 tidak ditolak.

Kesimpulan :
Berdasarkan hipotesis maka menunjukkan bahwa tidak ada autokorelasi yang signifikan dalam residual model, baik autokorelasi positif maupun negatif.

3. Uji Normalitas :

Hipotesis:

H0 (Hipotesis Nol): Residual terdistribusi normal.

H1 (Hipotesis Alternatif): Residual tidak terdistribusi normal.

#Perhitungan Variance Inflation Factor (VIF) untuk Mendeteksi Multikolinearitas

# Uji Spasial dan Estimasi Model Regresi Spasial dengan OLS

```python

!pip install pysal==2.6.0
!pip install esda
!pip install libpysal

import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import libpysal as ps
import spreg
import geopandas as gpd
from sklearn.preprocessing import StandardScaler


data = pd.read_excel('/content/data_kapsel.xlsx', nrows=36)
gdf = gpd.read_file("/content/BATAS KABUPATEN KOTA DESEMBER 2019 DUKCAPIL.shp")

# Pastikan 'Wilayah' di 'data' dan 'KAB_KOTA' di 'gdf' memiliki tipe data yang sama
data['Wilayah'] = data['Wilayah'].astype(str)
gdf['KAB_KOTA'] = gdf['KAB_KOTA'].astype(str)

# Gabungkan data berdasarkan 'Wilayah'
gdf.rename(columns={'KAB_KOTA': 'Wilayah'}, inplace=True)
merged_data = pd.merge(gdf, data, on='Wilayah', how='inner')

# Standarisasi data
scaler = StandardScaler()
standardized_cols = ['y', 'x1', 'x2', 'x3', 'x4', 'x5']
merged_data[standardized_cols] = scaler.fit_transform(merged_data[standardized_cols])

# Buat matriks bobot spasial Queen
w = ps.weights.Queen.from_dataframe(merged_data)
w.transform = 'r'

# Definisikan variabel dependen dan independen
y = merged_data['y'].values.reshape(-1, 1)
X = merged_data[['x1', 'x2', 'x3', 'x4', 'x5']].values

# Bangun model OLS spasial dengan spreg
m1 = spreg.OLS(
    y, X,
    w=w,
    spat_diag=True,
    name_x=merged_data[['x1', 'x2', 'x3', 'x4', 'x5']].columns.tolist(),
    name_y='y'
)

# Tampilkan ringkasan hasil model
print(m1.summary)
```

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
import statsmodels.api as sm
import libpysal as ps

# Hapus baris dengan NaN
merged_data_clean = merged_data.dropna(subset=['y', 'x1', 'x2', 'x3', 'x4', 'x5'])

# Standarisasi data
scaler = StandardScaler()

# Standarisasi y dan X
y_std = scaler.fit_transform(merged_data_clean[['y']]).flatten()
X_std = scaler.fit_transform(merged_data_clean[['x1', 'x2', 'x3', 'x4', 'x5']])

# Menambahkan konstanta ke X
X_std = sm.add_constant(X_std)

# Hitung spatial weights matrix (Queen's Contiguity)
W = ps.weights.Queen.from_dataframe(merged_data_clean)
W.transform = 'r'

# Menghitung spatial lag dari y
W_matrix = W.full()[0]
Wy = W_matrix @ y_std

# Gabungkan X dengan Spatial Lag (Wy)
X_spatial_std = np.column_stack([X_std, Wy])

# Cek dimensi
print("Dimensi X_spatial_std:", X_spatial_std.shape)
print("Dimensi y_std:", y_std.shape)

# Fit Robust Regression (Huber T)
model = sm.RLM(y_std, X_spatial_std, M=sm.robust.norms.HuberT())
results = model.fit()

# Tambahkan perhitungan nilai p-value
variables = ['Constant', 'x1', 'x2', 'x3', 'x4', 'x5', 'Spatial Lag']
coefficients = results.params
p_values = results.pvalues

# Gabungkan hasil ke dalam DataFrame
results_df = pd.DataFrame({
    'Variable': variables,
    'Coefficient': coefficients,
    'P-Value': p_values
})

# Tampilkan hasil lengkap
print("\nRobust Spatial Autoregressive Model Results:")
print("===========================================")
print(results_df)
```
